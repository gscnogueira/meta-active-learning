### Tentativa de encontrar os critérios

```python
datalist = openml.datasets.list_datasets(output_format="dataframe")
```

Com base nesse historgrama, foram utilizados os seguintes parâmetros de filtragem

```python
instances_condition = datalist.NumberOfInstances <= 10000

classes_condition = datalist.NumberOfClasses <= 10

numeric_features_condition = datalist.NumberOfNumericFeatures <= 500

symbolic_features_condition = datalist.NumberOfSymbolicFeatures <= 40

majority_class_condition =  (datalist.MajorityClassSize/datalist.NumberOfInstances) <= 1

minority_class_condition = (datalist.MinorityClassSize/datalist.NumberOfInstances) <= 0.5

datasets = datalist.loc[instances_condition & classes_condition &
                        numeric_features_condition & symbolic_features_condition &
                        majority_class_condition & minority_class_condition]

datasets
```

```python
print(f'Foram encontrados {len(datasets)} datasets que satisfazem essas condições')
```
